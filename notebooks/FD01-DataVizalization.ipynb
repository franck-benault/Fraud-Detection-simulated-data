```python
import seaborn as sns
import matplotlib.pyplot as plt

dfTemp=dfLearning.drop(['TRANSACTION_ID', 'TX_DATETIME', 'CUSTOMER_ID', 'TERMINAL_ID','TX_FRAUD_SCENARIO'],
  axis='columns', inplace=False)

commonCols= ['TX_AMOUNT', 'TX_TIME_SECONDS', 'TX_TIME_DAYS', 
       'TX_DURING_WEEKEND', 'TX_DURING_NIGHT']

#dfTemp=dfTemp.drop(commonCols,  axis='columns', inplace=False)

customerCols = [
    'CUSTOMER_ID_NB_TX_1DAY_WINDOW', 'CUSTOMER_ID_AVG_AMOUNT_1DAY_WINDOW',
       'CUSTOMER_ID_NB_TX_7DAY_WINDOW', 'CUSTOMER_ID_AVG_AMOUNT_7DAY_WINDOW',
       'CUSTOMER_ID_NB_TX_30DAY_WINDOW', 'CUSTOMER_ID_AVG_AMOUNT_30DAY_WINDOW']

dfTemp=dfTemp.drop(customerCols,  axis='columns', inplace=False)

terminalCols= ['TERMINAL_ID_NB_TX_1DAY_WINDOW', 'TERMINAL_ID_RISK_1DAY_WINDOW',
       'TERMINAL_ID_NB_TX_7DAY_WINDOW', 'TERMINAL_ID_RISK_7DAY_WINDOW',
       'TERMINAL_ID_NB_TX_30DAY_WINDOW', 'TERMINAL_ID_RISK_30DAY_WINDOW']

dfTemp=dfTemp.drop(terminalCols,  axis='columns', inplace=False)

matrix = dfTemp.corr()

plt.figure(figsize=(8,6))
sns.heatmap(matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Correlation Heatmap")
plt.show()
```
