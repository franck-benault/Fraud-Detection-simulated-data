# PCA

```python
#%%script false

from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from imblearn.under_sampling import RandomUnderSampler

predictors = dm.getPredictors(dfLearning)
target = dm.getTarget()

x1, y1 = dfLearning[predictors], dfLearning[target]

pca = PCA(n_components=2)
sc = StandardScaler()
x2 = sc.fit_transform(x1)
x_pca = pca.fit_transform(x2)

# giving a larger plot
plt.figure(figsize=(8, 6))

plt.scatter(x_pca[:, 0], x_pca[:, 1],
            c=y1,
           cmap='cool')

# labeling x and y axes
plt.xlabel('First Principal Component')
plt.ylabel('Second Principal Component')
plt.show()
```

```python
#%%script false

from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from imblearn.under_sampling import RandomUnderSampler

predictors = dm.getPredictors(dfLearning)
target = dm.getTarget()

undersample = RandomUnderSampler(sampling_strategy=0.01,random_state=42)
x1, y1 = undersample.fit_resample(dfLearning[predictors], dfLearning[target])

pca = PCA(n_components=2)
sc = StandardScaler()
x2 = sc.fit_transform(x1)
x_pca = pca.fit_transform(x2)

# giving a larger plot
plt.figure(figsize=(8, 6))

plt.scatter(x_pca[:, 0], x_pca[:, 1],
            c=y1,
            cmap='cool')
      #      cmap='plasma')

# labeling x and y axes
plt.xlabel('First Principal Component')
plt.ylabel('Second Principal Component')
plt.show()
```

# TSNE

```python
#%%script false
# computationally intensive
# very slow too slow 

from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from imblearn.under_sampling import RandomUnderSampler

predictors = dm.getPredictors(dfLearning)
target = dm.getTarget()

undersample = RandomUnderSampler(sampling_strategy=0.01,random_state=42)
x1, y1 = undersample.fit_resample(dfLearning[predictors], dfLearning[target])


sc = StandardScaler()
tsne = TSNE(n_components=2, random_state=42)
x2 = sc.fit_transform(x1)
x_tsne = tsne.fit_transform(x2)

# giving a larger plot
plt.figure(figsize=(8, 6))

plt.scatter(x_tsne[:, 0], x_tsne[:, 1],
            c=y1,
            cmap='cool')

# labeling x and y axes
plt.xlabel('First Principal Component')
plt.ylabel('Second Principal Component')
plt.show()
```
