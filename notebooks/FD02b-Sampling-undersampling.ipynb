#  PCA

```python


from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from imblearn.under_sampling import RandomUnderSampler

predictors = dm.getPredictors(dfLearning)
target = dm.getTarget()

x1, y1 = dfLearning[predictors], dfLearning[target]
sc = StandardScaler()
x2 = sc.fit_transform(x1)
print("dfLearning ",dfLearning.shape)
print("x2         ",x2.shape)


pca = PCA(n_components=2)
x_pca = pca.fit_transform(x2)

# giving a larger plot
plt.figure(figsize=(8, 6))

plt.scatter(x_pca[:, 0], x_pca[:, 1],
            c=y1,
           cmap='cool')

# labeling x and y axes
plt.xlabel('First Principal Component')
plt.ylabel('Second Principal Component')
plt.show()
```

# Random undersampling + PCA

```python
#%%script false

from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from imblearn.under_sampling import RandomUnderSampler
from datetime import datetime

predictors = dm.getPredictors(dfLearning)
target = dm.getTarget()

x1, y1 = dfLearning[predictors], dfLearning[target]
sc = StandardScaler()
x2 = sc.fit_transform(x1)

then= datetime.now()
undersample = RandomUnderSampler(sampling_strategy=0.01,random_state=42)
x3, y3 = undersample.fit_resample(x2,y1)
now = datetime.now()
duration= now - then
duration_in_s = duration.total_seconds()
print("1-Duration ",duration_in_s)
print("dfLearning ",dfLearning.shape)
print("x3         ",x3.shape)

pca = PCA(n_components=2)
x_pca = pca.fit_transform(x3)

# giving a larger plot
plt.figure(figsize=(8, 6))

plt.scatter(x_pca[:, 0], x_pca[:, 1],
            c=y3,
            cmap='cool')
      #      cmap='plasma')

# labeling x and y axes
plt.xlabel('First Principal Component')
plt.ylabel('Second Principal Component')
plt.show()
```
